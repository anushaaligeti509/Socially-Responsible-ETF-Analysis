### Processing

#### 2) AUM cleaning:
- Since all the values of the AUM variable don't have the same units, we change all to one unit i.e to millions
- Some of the values in the column are NaN and they don't have to be changed
- We create a function called `billion_to_million` to convert all the billion values to million
- The million and NaN values remain the same
- To find out the billion values we use regular expressions to create a pattern and find billion values and convert them to millions

```python
import re #import regular expressions library
df['AUM'] = df['AUM'].astype(str).str.replace('[\$]', '', regex=True) #replace dollar sign with ''
df['AUM']
```

```python
#defining a function which turns billions to millions

def billion_to_million(s): 
    if s == np.NaN:         #if value is NaN keep it same
        result = np.NaN
    else:
        pattern = r'(\d+\.\d+)B'   #regular expression pattern for billion
        result = re.sub(pattern, lambda match: str(float(match.group(1)) * 1000),str(s))
    return result
```

- Applying the function onto the `AUM` column of the dataframe to convert the billion values

```python
df['AUM'] = df['AUM'].apply(billion_to_million)  #appying the function to AUM
```

- As the AUM column contains all millions now, but there are units from the original million values, strip the million sign from the dataframe and replace M with ''
- For the Further analysis, we may have to work with AUM data, so we convert the column to numeric so we can perform operations on it

```python
df['AUM'] = df['AUM'].astype(str).str.replace('[M]', '', regex=True) #replacing M with ''
df['AUM'] = pd.to_numeric(df['AUM'], errors = 'coerce') #converting the AUM column into numeric data
```

- By applying the function onto the column 'AUM', all the values of the column had to be converted to datatype `str`
- The NaN values also turn into strings so we replace them back to np.NaN

```python
df = df.replace('nan', np.nan)   #replacing nan to NaN
df
```

#### 3) Splitting the Sement Column
- By studying the segment column, we notice that `:` seperates the asset class with the other market and segment
- We split the Asset class by `:` and place it into a new column and the rest of the split data to a segment column

```python
#splitting segment into segment and asset class
df[['Asset Class','ETF Segment']] = df['Segment'].str.split(':', expand=True)
```

- ` - `seperates the market with the segment, so we split the string and place it into a new column market

```python
#splitting segment again to market and segment
df[['Market','ETF Segment']] = df['ETF Segment'].str.split(' - ', n=1, expand=True)
```

#### 10) Frequency distributions of Asset Class, markets and segments for funds of each issuer
